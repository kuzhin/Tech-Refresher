```python
import pandas as pd
import numpy as np
```

## Задача: Дан дф, нужно написать решение, в результате выполнения кода, останутся id с температурой выше, чем в предыдущий день

```python
data = {
    'id': [1, 2, 3, 4],
    'recordDate': ['2015-01-01', '2015-01-02', '2015-01-03', '2015-01-04'],
    'temperature': [10, 25, 20, 30]
}
df = pd.DataFrame(data)
print(df)
print(data['recordDate'])
print(df['recordDate'].dtype)
```

### Решение:

```python
# Подготовка дф для анализа:
# привели даты к нужному формату, для дальнейшей сортировки
df['recordDate'] = pd.to_datetime(df['recordDate'])
df = df.sort_values(by='recordDate')

# выполнение самого задания:
df['prev_temp'] = df['temperature'].shift(1)
df['prev_date'] = df['recordDate'].shift(1)
df = df[
    (df['prev_temp'] < df['temperature']) &
    (df['recordDate'] == df['prev_date'] + pd.Timedelta(days=1))
][['id']].rename(columns={'id': 'Id'})
print(df)
```
